```python
import pandas as pd
from preprocessing_utilites import NumericalPreprocessingToolkit

# Load the dataset
df = pd.read_csv(
    './data/airbnb/Airbnb Data/Listings.csv', encoding='ISO-8859-1')

# Show a preview of the data to verify
df.head()
```

```python
columns_to_process = [
    'accommodates', 'bedrooms', 'price', 'minimum_nights', 'maximum_nights',
    'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
    'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
    'review_scores_value'
]

toolkit = NumericalPreprocessingToolkit(df)

df_missing_handled = toolkit.handle_missing_values(
    columns=columns_to_process, method='mean')
print("DataFrame after Handling Missing Values:")
df_missing_handled.head()
```

```python
df_standardized = toolkit.standardize(columns=columns_to_process)
print("\nDataFrame after Standardization:")
df_standardized.head()
```

```python

df_normalized = toolkit.normalize(
    columns=columns_to_process, feature_range=(0, 1))
print("\nDataFrame after Normalization:")
df_normalized.head()
```

```python
from preprocessing_utilites import CategoricalUtilityToolkit

# Initialize the toolkit with the DataFrame
cat_toolkit = CategoricalUtilityToolkit(df_normalized)

# List of categorical columns we want to process
categorical_columns = [
    'name', 'host_location', 'host_response_time', 'host_is_superhost',
    'host_has_profile_pic', 'host_identity_verified', 'neighbourhood',
    'district', 'city', 'property_type', 'room_type'
]

# 1. Handling Missing Values in Categorical Columns
df_no_missing = cat_toolkit.handle_missing(columns=categorical_columns)
print("DataFrame after handling missing values:")
df_no_missing[categorical_columns].head()
```

```python

# 2. Label Encoding for Categorical Columns
# We'll label encode columns with few distinct categories, such as `room_type` and `host_is_superhost`.
label_encode_columns = ['room_type', 'host_is_superhost']
df_label_encoded = cat_toolkit.label_encode(columns=label_encode_columns)
print("\nDataFrame after label encoding:")
df[label_encode_columns] = df_label_encoded[label_encode_columns]
df_label_encoded[label_encode_columns].head()
```

```python

# 3. One-Hot Encoding for Categorical Columns
# We'll apply one-hot encoding to multi-category columns such as `city` and `property_type`.
one_hot_columns = ['city', 'property_type']
df_one_hot_encoded = cat_toolkit.one_hot_encode(columns=one_hot_columns)
print("\nDataFrame after one-hot encoding:")
df = df_one_hot_encoded
df_one_hot_encoded.head()
```

```python

# 4. Handling Rare Categories
# We handle rare categories in columns like `host_location` and `neighbourhood`.
rare_category_columns = ['host_location', 'neighbourhood']
df_rare_handled = cat_toolkit.handle_rare_categories(
    columns=rare_category_columns, threshold=0.05)
print("\nDataFrame after handling rare categories:")
df[rare_category_columns] = df_rare_handled[rare_category_columns]
df_rare_handled[rare_category_columns].head()
```

```python
from preprocessing_utilites import NLPUtilityToolkit
nlp_toolkit = NLPUtilityToolkit()

# Apply various NLP operations
df = nlp_toolkit.tokenize(df_rare_handled, 'name')
df.head()
```
